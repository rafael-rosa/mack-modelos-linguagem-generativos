```python
import pandas as pd
import os
```

```python
datasets_path = "..\\coleta_dados\\out"

onlyfiles = [os.path.join(dirpath,f) for (dirpath, dirnames, filenames) in os.walk(datasets_path) for f in filenames]

print(len(onlyfiles))
print(onlyfiles)

df = pd.concat(map(pd.read_csv, onlyfiles), ignore_index=True)

df.shape
```

```python
df['title'] = df['title'].str.upper()
df['rated'] = df['rated'].str.upper()
df['genre'] = df['genre'].str.upper()
df['director'] = df['director'].str.upper()
df['writer'] = df['writer'].str.upper()
df['actors'] = df['actors'].str.upper()
df['plot'] = df['plot'].str.upper()
df['language'] = df['language'].str.upper()
df['country'] = df['country'].str.upper()
df['awards'] = df['awards'].str.upper()

df.head()
```

```python
# remover registros com coluna plot e genre vazias
df = df.dropna(subset=['plot', 'genre'])
df = df.reset_index(drop=True)

df.shape

```

```python
# remover colunas desnecessárias
# df = df.drop(columns=['metascore', 'country', 'awards', 'language', 'imdb_id'])
# manter apenas plot e genre
df = df[['plot', 'genre']]

# manter apenas filmes com apenas 1 gênero
df = df[df['genre'].str.count(',') == 0]
df = df.reset_index(drop=True)
```
